```python
import numpy as np
from astropy.io import fits

data_dir = '../../data/'

hdulist = fits.open(data_dir+'boss_skyrenorm_v5_13_0.fits')
hdu = hdulist[1]
```

```python
#https://docs.astropy.org/en/stable/table/
#https://docs.astropy.org/en/stable/io/unified.html#table-io
from astropy.table import Table
t = Table.read(data_dir+'boss_skyrenorm_v5_13_0.fits', format='fits')
```

```python
t
```

```python
t.info
```

```python
skyloglam = t['SKYLOGLAM'][0]
skyrms = t['SKYRMS'][0]
skyrmsfit = t['RMSFIT'][0]
skyrenorm = t['SKYRENORM'][0]
smask = t['SMASK'][0]

from matplotlib import pyplot as plt
import sys
sys.path.append('../')
import utils as ul

font = {'family':'serif', 'size':15}
plt.rc('font', **font)
plt.rc('xtick', labelsize='small')
plt.rc('ytick', labelsize='small')
plt.rc('axes', labelsize='medium')

fig,ax = plt.subplots()
ax.plot(10**skyloglam,skyrms,linewidth=1.0)
ax.plot(10**skyloglam,skyrmsfit,'red')
id_mask = ul.return_mask_region_id(smask)
for this_mask_id in id_mask:
    low = this_mask_id[0]
    high = this_mask_id[1]
    ax.axvspan(10**skyloglam[low], 10**skyloglam[high-1], facecolor='g', alpha=0.5)
ax.plot(10**skyloglam,skyrenorm,'--',alpha=0.7,linewidth=0.7)
ax.minorticks_on()
ax.set_xlim(3700,10000)
ax.set_ylim(0.,4.0)
ax.set_xlabel(r'Observed Wavelength $[\AA]$') #fontsize=20
ax.set_ylabel(r'$\chi(\lambda)$ / Recaling factor')
fig.savefig('./noise_rescaling.pdf',bbox_inches='tight')
```
