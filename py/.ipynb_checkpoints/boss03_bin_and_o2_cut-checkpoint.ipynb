```python
import subprocess
import os
import utils as ul
import copy
import pickle
import gzip
from matplotlib import pyplot as plt
import numpy as np
from scipy.signal import medfilt

dir_data = '/data/inspur_disk03/userdir/caoxy/eboss_lya/data/'
lya_hits = pickle.load(gzip.open(dir_data+'lya_hits.pzip'))

obswave = lya_hits['peak_wave']
restwave = obswave/(1.0+lya_hits['z_hits'])

print('number of lya hits is:',lya_hits['fiber_hits'].size)
plt.figure()
plt.scatter(np.log10(obswave),np.log10(restwave),s=0.5)
plt.xlabel('log10(obswave/Ang)')
plt.ylabel('log10(restwave/Ang)')
plt.title('One-line hits')
plt.show()

#---------------------------cut on rest-frame
dloglam = 3. * 0.0001 #three times the SDSS redux pixel unit
minloglam = np.min(np.log10(restwave))
maxloglam = np.max(np.log10(restwave))
nbins = np.ceil((maxloglam - minloglam) / dloglam)
restbase = dloglam * (np.arange(nbins) + 0.5) + minloglam
bins = dloglam * np.arange(nbins+1) + minloglam
resthist = np.histogram(np.log10(restwave),bins=bins)[0] 
#the 0-element is hist number, 1-elements is bins

#Wide median filter:
histwid = 50
hist_val = medfilt(resthist, 2*histwid+1)
hist_val[0]=resthist[0];hist_val[-1]=resthist[-1] #compatibility to idl
histfilt = np.clip(hist_val,1.0,None)


histfilt[0:histwid+1] = histfilt[histwid+1]
histfilt[int(nbins)-histwid-1:] = histfilt[int(nbins)-histwid-2]

sigthresh = 4.
histmask = (resthist > (histfilt + sigthresh*np.sqrt(histfilt)))

plt.figure()
plt.bar(restbase, resthist,alpha=0.5,color='blue',width=restbase[1]-restbase[0])
plt.bar(restbase, histfilt,alpha=0.5,color='orange',width=restbase[1]-restbase[0])
plt.bar(restbase, resthist*(1-histmask),alpha=0.5,color='red',width=restbase[1]-restbase[0])
plt.xlabel('bins')
plt.ylabel('number of hits')
plt.title('Target galaxy rest-frame hitwave histogram')
plt.show()


#Grow the mask a bit:
ngrow = 1
histmask = ul.smooth(histmask.astype('float'), 2*ngrow+1) > 0.0

#Figure out who gets cut based on this mask:
cutit = np.interp(np.log10(restwave), restbase, histmask.astype('float')) > 0.5
id_rest = np.where(cutit == 0)

nhits_1 = id_rest[0].size
lya_hits1 = ul.cut_dict(lya_hits,id_rest)
print('number of lya hits is:',lya_hits1['fiber_hits'].size)

#show it works
obswave = lya_hits1['peak_wave']
restwave = obswave/(1.0+lya_hits1['z_hits'])
plt.figure()
plt.scatter(np.log10(obswave),np.log10(restwave),s=0.5)
plt.xlabel('log10(obswave/Ang)')
plt.ylabel('log10(restwave/Ang)')
plt.title('One-line hits')
plt.show()

#---------------------------cut on obs-frame
dloglam = 3. * 0.0001 #three times the SDSS redux pixel unit
minloglam = np.min(np.log10(obswave))
maxloglam = np.max(np.log10(obswave))
nbins = np.ceil((maxloglam - minloglam) / dloglam)
obsbase = dloglam * (np.arange(nbins) + 0.5) + minloglam
bins = dloglam * np.arange(nbins+1) + minloglam
obshist = np.histogram(np.log10(obswave),bins=bins)[0] 
#the 0-element is hist number, 1-elements is bins

#Wide median filter:
histwid = 50
hist_val = medfilt(obshist, 2*histwid+1)
hist_val[0]=obshist[0];hist_val[-1]=obshist[-1] #compatibility to idl
histfilt = np.clip(hist_val,1.0,None)

histfilt[0:histwid+1] = histfilt[histwid+1]
histfilt[int(nbins)-histwid-1:] = histfilt[int(nbins)-histwid-2]

sigthresh = 4.
histmask = (obshist > (histfilt + sigthresh*np.sqrt(histfilt)))

plt.figure()
plt.bar(obsbase, obshist,alpha=0.5,color='blue',width=obsbase[1]-obsbase[0])
plt.bar(obsbase, histfilt,alpha=0.5,color='orange',width=obsbase[1]-obsbase[0])
plt.bar(obsbase, obshist*(1-histmask),alpha=0.5,color='red',width=obsbase[1]-obsbase[0])
plt.xlabel('bins')
plt.ylabel('number of hits')
plt.title('Target galaxy obs-frame hitwave histogram')
plt.show()

#Grow the mask a bit:
ngrow = 1
histmask = ul.smooth(histmask.astype('float'), 2*ngrow+1) > 0.0

#Figure out who gets cut based on this mask:
cutit = np.interp(np.log10(obswave), obsbase, histmask.astype('float')) > 0.5
id_obs = np.where(cutit == 0)

nhits_2 = id_obs[0].size
lya_hits2 = ul.cut_dict(lya_hits1,id_obs)
print('number of lya hits is:',lya_hits2['fiber_hits'].size)

#show it works
obswave = lya_hits2['peak_wave']
restwave = obswave/(1.0+lya_hits2['z_hits'])
plt.figure()
plt.scatter(np.log10(obswave),np.log10(restwave),s=0.5)
plt.xlabel('log10(obswave/Ang)')
plt.ylabel('log10(restwave/Ang)')
plt.title('One-line hits')
plt.show()


#oII cut-----------------------------------------
sn_oii = 2.5
is_o2 = np.sum(lya_hits2['fakesn']>sn_oii, axis=1) > 0
id_cut_oii = np.where(is_o2==0)

nhits_3 =id_cut_oii[0].size
lya_hits3 = ul.cut_dict(lya_hits2,id_cut_oii)
print('number of lya hits is:',lya_hits3['fiber_hits'].size)

#show it works
obswave = lya_hits3['peak_wave']
restwave = obswave/(1.0+lya_hits3['z_hits'])
plt.figure()
plt.scatter(np.log10(obswave),np.log10(restwave),s=0.5)
plt.xlabel('log10(obswave/Ang)')
plt.ylabel('log10(restwave/Ang)')
plt.title('cut Oii hits')
plt.show()

#save the final hits
out_dir = '/data/inspur_disk03/userdir/caoxy/eboss_lya/data/'
#pickle.dump(lya_hits3,gzip.open(out_dir+'lya_hits_cut.pzip','wb'))
```

```python
import numpy as np
a=np.random.rand(5)*10
print(a)
k_size = 3
kernal = np.ones(k_size)/k_size
b=np.correlate(a,kernal,mode='same')
print(b)
#in idl smooth fucntion,the 0 and -1 elements of b are set to a's value
```

```python
def smooth(array,kernal_size):
    """
    simulate the idl smooth function
    """
    kernal = np.ones(kernal_size)/kernal_size
    res = np.correlate(array,kernal,mode='same')
    res[0] = array[0]
    res[-1] = array[-1]
    return res
```
